# Inicio filtrado y preparación de datos

## Importo las librerías

```python
import pandas as pd 
```

## Cargo las bases de datos

```python
usua = pd.read_csv("../data/raw/usuarios_ecobici_2024.csv", sep=',', encoding='utf_8')

reco = pd.read_csv("../data/raw/badata_ecobici_recorridos_realizados_2024.csv", sep=',', encoding='utf_8')
```

Observo que los nombres de las columnas ya están en formato snake_case, por lo tanto en ese paso no realizo ninguna acción

# Convierto fechas a tipo datetime

En la base Recorridos realizados 2024 hay 2 columnas: fecha_origen_recorrido y fecha_destino_recorrido, las cambio

```python
reco['fecha_destino_recorrido'] = pd.to_datetime(reco['fecha_destino_recorrido'])
reco['fecha_origen_recorrido'] = pd.to_datetime(reco['fecha_origen_recorrido'])
```

Realizo un info para comprobar que se cambiaron

```python
reco.info()
```

En la base Usuarios Ecobici 2024 hay solamente 1 columna: fecha_alta, la cambio y realizo un info para comprobar

```python
usua['fecha_alta'] = pd.to_datetime(usua['fecha_alta'])

usua.info()
```

# Filtrado de los datos de julio, agosto y septiembre 2024

Para esto utilizo 2 variables, una para el primer día de julio y otra para el último día de septiembre, y realizo el filtrado de la base de datos de recorridos

```python
# filtrar los viajes de julio, agosto y septiembre
fecha_inicio = '2024-07-01'
fecha_fin = '2024-09-30'
recorridos_3meses = reco[(reco['fecha_origen_recorrido'] >= pd.to_datetime(fecha_inicio)) 
                         & (reco['fecha_destino_recorrido'] <= pd.to_datetime(fecha_fin))]
```

```python
recorridos_3meses
```

Encuentro que durante esos 3 meses se realizaron 817.631 recorridos

## Guardo este resultado intermedio en data/interim/recorridos_3meses.csv

```python
ruta_guardado = "../data/interim/recorridos_3meses.csv"
recorridos_3meses.to_csv(ruta_guardado, index=False)
```
